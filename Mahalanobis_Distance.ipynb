Consider a random variable $w=({w_1},{w_2})\in{\mathbb{R}}^2$ where we know:
$${\mu}_w=E[w]=\begin{bmatrix} 1 \\ 2 \end{bmatrix},\quad{\Sigma}_w=Var[w]=\begin{bmatrix} 6 & -4 \\ -4 & 3\end{bmatrix}$$

We will make a clear graph showing the set of samples from the distribution of $w$ which have a Mahalanobis distance equal to **1**. In the sense of the Mahalanobis distance, these are the samples that are all equally consistent with the known statistics of $w$. 

```python
import matplotlib.pyplot as plt
import numpy as np

w1_samp = np.linspace(1 - np.sqrt(6), 1 + np.sqrt(6), 100)
w2_samp1 = -2*w1_samp/3 - np.sqrt(-2*w1_samp**2 + 4*w1_samp + 10)/6 + 8/3
w2_samp2 = -2*w1_samp/3 + np.sqrt(-2*w1_samp**2 + 4*w1_samp + 10)/6 + 8/3

plt.figure(figsize = (8, 8))

plt.plot(w1_samp, w2_samp1, 'b-', w1_samp, w2_samp2, 'b-')
plt.scatter(1, 2, label = 'Expectation Point')

plt.legend()
plt.xlabel(r'$w_1$', fontsize = 15)
plt.ylabel(r'$w_2$', fontsize = 15)
plt.title('Samples With Mahalanobis Distance of 1', fontsize = 15)

plt.show()
```
